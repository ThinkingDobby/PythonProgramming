```python
from math import gcd

n = int(input())
f = list(map(int, input().split()))
m = int(input())
s = list(map(int, input().split()))
a = f[0]
b = s[0]
for i in range(1, n):
    a *= f[i]
for j in range(1, m):
    b *= s[j]

tmp = str(gcd(a, b))
print(tmp[max(0,len(tmp) - 9)::])
```
