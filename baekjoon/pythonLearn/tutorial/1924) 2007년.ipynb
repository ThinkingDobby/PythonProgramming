```python
x, y = map(int, input().split())
memo = [0 for _ in range(0, 13)]
for i in range(1, 13):
    if i in [1, 3, 5, 7, 8, 10, 12]:
        memo[i] += memo[i - 1] + 31
    elif i == 2:
        memo[i] += memo[i - 1] + 28
    else:
        memo[i] += memo[i - 1] + 30

days = ["MON", "TUE", "WED", "THU", "FRI", "SAT", "SUN"]
print(days[(memo[x - 1] + y - 1) % 7])
```
