```python
t = int(input())
for i in range(t):
    n = input().split()
    ans = 0
    if n[1] == '+':
        ans = int(n[0]) + int(n[2])
    elif n[1] == '-':
        ans = int(n[0]) - int(n[2])
    elif n[1] == '*':
        ans = int(n[0]) * int(n[2])
    else:
        ans = int(n[0]) // int(n[2])

    if ans == int(n[4]):
        print("correct")
    else:
        print("wrong answer")
```
