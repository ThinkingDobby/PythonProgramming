```python
comb = [[j for j in range(0, 101)] for i in range(0, 101)]
comb[1][0] = 1
comb[1][1] = 1

for i in range(2, 101):
    for j in range(0, i + 1):
        if (j == 0 or j == i):
            comb[i][j] = 1
        else:
            comb[i][j] = comb[i - 1][j - 1] + comb[i - 1][j]

n, m = map(int, input().split())
print(comb[n][m])
```

```python
from math import factorial as f

n, m = map(int, input().split())
print(f(n) // f(m) // f(n - m))
```

```python
from math import comb

n, m = map(int, input().split())
print(comb(n, m))
```
