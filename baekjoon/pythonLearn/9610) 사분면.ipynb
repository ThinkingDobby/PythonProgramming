```python
memo = [0, 0, 0, 0, 0]

for _ in range(int(input())):
    x, y = map(int, input().split())
    if x > 0 and y > 0: memo[0] += 1
    elif x < 0 and y > 0: memo[1] += 1
    elif x < 0 and y < 0: memo[2] += 1
    elif x > 0 and y < 0: memo[3] += 1
    else: memo[4] += 1

for i in range(4):
    print("Q%d: %d"%(i + 1, memo[i]))
print("AXIS: %d" %memo[4])
```
