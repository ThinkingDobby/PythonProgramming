```python
def check_p(num):
    check = {i for i in range(1, num // 2 + 1) if num % i == 0}
    return num == sum(check), map(str, sorted(list(check)))

while True:
    iv = int(input())
    if iv == -1: break
    check = check_p(iv)
    if check[0]:
        print(iv, "=", " + ".join(check[1]))
    else:
        print(iv, "is NOT perfect.")
```
