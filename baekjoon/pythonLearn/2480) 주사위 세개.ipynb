```python
n = list(map(int, input().split()))
if n[0] == n[1] and n[1] == n[2]:
    print(n[0] * 1000 + 10000)
elif n[0] == n[1] or n[1] == n[2]:
    print(n[1] * 100 + 1000)
elif n[0] == n[2]:
    print(n[0] * 100 + 1000)
else:
    print(max(n) * 100)
```

```python
# 참조한 코드
l=sorted(list(map(int,input().split(' '))))
if len(set(l))==1:
    print(10000+(l[0]*1000))
elif len(set(l))==2:
    print(1000+l[1]*100)    # l[1] 값은 항상 겹치는 값
else:
    print(max(l)*100)
```
