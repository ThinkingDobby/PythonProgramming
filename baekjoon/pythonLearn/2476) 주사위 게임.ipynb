```python
m = -1
for _ in range(int(input())):
    a = list(map(int, input().split()))
    a.sort()
    if a[0] == a[2]:
        m = max(m, 10000 + a[0] * 1000)
    elif a[0] == a[1] or a[1] == a[2]:
        m = max(m, 1000 + a[1] * 100)
    else:
        m = max(m, a[2] * 100)

print(m)
```
