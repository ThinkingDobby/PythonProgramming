```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
plt.plot(x)
plt.show()
```

```python
y = [13, 16, 15, 18, 16]
plt.plot(x, y)
plt.show()
```

```python
plt.plot([1, 2, 3], [4, 5, 6], color="red", linewidth=5)
plt.plot(2, 2, 'bo')
plt.plot(2, 6, 'g*')
plt.show()
```

```python
x = [chr(i) for i in range(97, 102)]
y = [1, 2, 3, 4, 5]
plt.plot(x, y, 'r--')
plt.xlabel("A")
plt.ylabel("B")
plt.show()
```

```python
x = [i for i in range(-10, 21)]
y = [2 * i - 5 for i in range(-10, 21)]

plt.plot(x, y, "r--")
plt.plot(x, x, "b")
plt.show()
```

```python
import math

f1 = lambda x: math.e**(x - 2) + 2
f2 = lambda x: (x - 2)**2

x = [i / 10 for i in range(0, 101)]
y1 = list(map(f1, x))
y2 = list(map(f2, x))

plt.plot(x, y1)
plt.plot(x, y2)

plt.scatter(x, y1)
plt.scatter(x, y2) 

plt.axis([-2, 11, -20, 20])

plt.show()
```

```python
import numpy as np

a = np.array([1, 2, 3, 8])
b = np.array([2, 5, 1, 2])

plt.plot(a, b)
plt.show()
```

```python
x = np.random.standard_normal(1000)
plt.hist(x)
plt.show()
```

```python
s = 100
x = np.random.rand(s)
y = list(map(lambda x: x + np.random.rand(), list(map(math.sin, x * 10))))
colors = np.random.rand(s)
shape = np.pi * (np.random.rand(s) * 10) ** 2

plt.scatter(x, y, s = shape, c = colors, marker = ".", alpha = 0.5)
plt.show()
```
