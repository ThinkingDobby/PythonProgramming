```python
import pandas as pd

x = pd.Series([7, 3, 5, 8], index=['서울', '대구', '부산', '광주'])
print(x)
print(x[['서울', '대구']])
```

```python
print(x.index)
print(x.values)
```

```python
x[sorted(x.index)]  # 인덱스로 정렬
```

```python
x[sorted(x.index, key=lambda a: x[a])]  # 값으로 정렬
```

```python
print(x.reindex(sorted(x.index)))
print(x)
```

```python
age = {'민준':23, '현우': 43, '서연':12, '동현': 45}    # 딕셔너리로 Series 생성 가능
x = pd.Series(age)
print(x)

print(x['동현'])    # 인덱스로 접근 가능
print(x[3]) # 인덱스에 해당하는 하나의 수로도 접근 가능
print(x['현우':'동현'])
# 하나의 수로 접근 시 값만 출력
print(x[0:2])   # 슬라이싱은 인덱스에 해당하는 수로 수행
# 슬라이싱으로 접근 시 인덱스, 값 모두 출력

x = pd.Series([1, 2, 3, 4], index=[2, 3, 4 ,5])
print(x)
print(x[2]) # 인덱스가 숫자인 경우에는 인덱스로만 접근 가능
print(x[0:2])   # 슬라이싱은 여전히 인덱스에 해당하는 수로 수행
print(x[1:2])


# 요약: 인덱스가 문자면 인덱스, 인덱스번호로 인덱싱, 슬라이싱 모두 가능 / 인덱스가 숫자면 인덱스로 인덱싱, 인덱스 번호로 슬라이싱
```

```python
from IPython.display import display

data = {'age':[23, 43, 12, 45],
    'name':['민준', '현우', '서연', '동현'],
    'height':[175.3, 180.3, 165.8, 172.7]}
x = pd.DataFrame(data)

display(x)
display(x.name)

print(x.iloc[0, [0, 1]])
print(x.iloc[0, 0:2])   # 슬라이싱, 튜플로 표현 동일
print(x.iloc[:, -1])
print(x.iloc[1:4, 1])
```
