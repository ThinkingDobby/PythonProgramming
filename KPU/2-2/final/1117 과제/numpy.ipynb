### 8번
참고</br>
element-wise: 원소 별</br>
complex number: 복소수</br>
</br>
파이썬에서 복소수 표현:</br>
i 대신 j 또는 J 붙임 → ex) 2 + 3j</br>
실수부, 허수부 구분해 표현 → ex) complex(2, 3)</br>


```python
import numpy as np

a = np.array([1 + 1j, 1 + 2j, 1.1, 2, 3.3])
print("주어진 배열:", a)
print("복소수 검사:", np.iscomplex(a))
print("실수 검사:", np.isreal(a))
print("스칼라 검사:", [np.isscalar(i) for i in a])
```

### 38번
참고</br>
numpy배열.tobytes(): 배열 -> 바이트 변환</br>
np.frombuffer(바꾸고 싶은 bytes , dtype = <자료형>): 바이트 -> 배열 변환

```python
import numpy as np

a = np.array([1, 2, 3])
print("주어진 배열:", a)
aBytes = a.tobytes()
print("바이트 변환:", aBytes)
aArr = np.frombuffer(aBytes, dtype = a.dtype)
print("배열 변환:", aArr)
```
