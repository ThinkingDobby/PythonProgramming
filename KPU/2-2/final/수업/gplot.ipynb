```python
import matplotlib.pyplot as plt

f = lambda x: 2 * x - 5
x = [i for i in range(-10, 21)]
y = list(map(f, x))

plt.plot(x, y, "r--")
plt.plot(x, x, "b")
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

f1 = lambda x: math.exp(1)**(x - 2) + 2
f2 = lambda x: (x - 2)**2

x = [i / 10 for i in range(0, 110)] # 그래프 자연스럽게 하기 위함
y1 = list(map(f1, x))
y2 = list(map(f2, x))

plt.title("relation of height and weight")

plt.plot(x, y1)
plt.plot(x, y2)

plt.xlabel("height")
plt.ylabel("weight")

plt.scatter(x, y1)  # 스캐터 차트
plt.scatter(x, y2)

plt.axis([-2, 11, -20, 20])
plt.show()

# 보완된 코드

# x = np.arange(0, 11, 0.1)

# plt.plot(x, np.e ** (x - 2) + 2)
# plt.plot(x, (x - 2)**2)

# plt.axis([-2, 11, -20, 20])
# plt.title("relation")

# plt.xlabel("height")
# plt.ylabel("weight")
```

```python
x = np.linspace(0, 5, 10)
y = x ** 2

print(x)

plt.plot(x, y)
plt.show()
```

```python
x = np.logspace(-1, 1, 20)
print(x)

y1 = x**2.0
y2 = x**1.5

plt.plot(x, y1, "bo-", linewidth=2, markersize=5, label="First")
plt.plot(x, y2, "gs-", linewidth=2, markersize=5, label="Second")
plt.hist(x)

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pylab as plt

s = 100
x = np.random.rand(s)
y = list(map(lambda x: x + np.random.rand(), list(map(math.sin, x * 10))))
colors = np.random.rand(s)
shape = np.pi * (np.random.rand(s) * 10) ** 2

plt.scatter(x, y, s=shape, c=colors, marker=".", alpha=0.3)
plt.show()
```
