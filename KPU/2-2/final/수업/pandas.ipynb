```python
import pandas as pd

x = pd.Series([i for i in range(10, 20, 2)], index = ['서울', '부산', '광주', '대전', '대구'])
print(x)
```

```python
x.reindex(sorted(x.index))  # 별도 저장 안하면 원본은 유지됨
```

```python
x = pd.Series([3, 8, 5, 9], index=['서울', '대구', '부산', '광주'])
y = pd.Series([2, 4, 5, 1], index=['대구', '부산', '서울', '대전'])
x + y   # 인덱스를 기준으로 연산
```

```python
medal = [1, 3, 2, 4, 2, 3]
x = pd.Series(medal)
print(pd.unique(x))
```

```python
x = pd.Series([3, 8, 5, 9], index=['서울', '대구', '부산', '광주'])
print(x[['서울', '부산']])  # 인덱스를 리스트 형태로 받아서 한번에 여러 요소 접근 가능
# 리스트 쓰지 않고 인덱스에 여러 요소 입력 시 오류 발생
print(x['서울', '대구'])
```

```python
print(x)
print(x[2:5])
print(x[3])
x[3] = 40
print(x)
x['부산'] = -40
print(x)
print(type(x.values))
print(x.values > 20)
print(x[x.values > 20])
print('대구' in x)  # x에 대해 in 키워드 사용 - index에 대해 조사
print(8 in x)
```

```python
x = pd.Series([11, 22, 33, 44])
x.reindex([3, 5, 1, 0], method="bfill")
```

```python
data = {'age' : [23, 43, 12, 45], 'name' : ['민준', '현우', '서연', '동현']
, 'height' : [175.3, 180.3, 165.8, 172.7]}
x = pd.DataFrame(data, columns=['name', 'age', 'height'], index=[1, 2, 3, 4])
x
```

```python
print(x['name'])
x.name  # 시리즈로 출력 
```

```python
print(x.iloc[0, [1, 2]])    # 1:3 과 [1, 2] 동일
print(x.iloc[0, 1:3])
print(x.iloc[[1, 3], [1, 2]])
```

```python
x[x.age > 30]
```

```python
x.mean(1)   # axis 설정 - 컬럼을 변화시키면서 구함(행 기준으로 결과 나옴)
# 숫자들에 대해서만 계산됨
```

```python
x.mean(0)    # 인덱스를 변화시키면서 구함(열 기준으로 결과 나옴)
```
