```python
import re

data = """
park 800905-1049118
kim  700905-1059119
"""

pat = re.compile("(\d{6})[-]\d{7}")
print(pat.sub("\g<1>-*******", data))
```

```python
import re

data = """
park 800905-1049118
kim  700905-1059119
"""

pat = re.compile("[1-9]+")
print(pat.match(data))
```

```python
p = re.compile("[a-z]+")

m = p.search("python")
print(m.group())
```

```python
m = re.match("[a-z]+", "python python")
print(m.group())
```

```python
p = re.compile("[a-z]+")

m = p.match("3 pyhton")
print(m)
m = p.search("3 python")    # search 는 전체를 대상으로
print(m.group())
```

```python
import re

p = re.compile("0\d{2}-\d{3,4}(?=-)-\d{4}(?=,)")
m = p.findall("010-1234-5678, 010-123-45678, 031-8041-0010, 123-456-7890")
print(m)
```

```python
import re
p = re.compile("^python\s\w+", re.M)

data = """python one
life is too short
python two
you need python
python three"""

print(p.findall(data))
```

```python
charref = re.compile(r"""
&[#]
(
    0[0-7]+
    | [0-9]+
    | x[0-9a-fA-F]+
)
;
""", re.VERBOSE)

print(charref.search("0123123"))
```

```python
p = re.compile('Crow|Servo')
m = p.match('CrowHello')
print(m)
```

```python
print(re.search('^Life', 'Life is too short'))
print(re.search('^Life', 'My Life'))
```

```python
p = re.compile("(ABC)+")
m = p.search("ABCABCABC ok?")
print(m.group())
```
