## Problem 28
수 1부터 시작해서 우측으로부터 시계방향으로 5×5 행렬을 만들면 아래와 같다.</br>
21 22 23 24 25</br>
20  7  8  9 10</br>
19  6  1  2 11</br>
18  5  4  3 12</br>
17 16 15 14 13</br>
1001x1001 행렬을 만들었을 때 대각선 상의 수를 더한 값을 구하시오.

```python
s = 1
now = 2
memo = 8
for i in range(3, 1002, 2):
    start = now + i - 2

    for j in range(0, 4):
        s += start + (i - 1) * j 

    now += memo
    memo += 8

print(s)
```

## Problem 56
a, b < 100 인 자연수 a와 b에 대해서, a^b의 자릿수의 합이 최대인 경우의 값을 출력하시오.

```python
m = 0
for a in range(1, 100):
    for b in range(1, 100):
        tmp = sum(map(int, str(a ** b)))
        if m < tmp: m = tmp

print(m)
```
