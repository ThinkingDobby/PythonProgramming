```python
import matplotlib.pyplot as plt

with open('file.txt', 'r') as f:
    x = [len(i.split()) for i in f.readlines()]
plt.bar(range(1, len(x) + 1), x)
plt.xlabel("line number")
plt.ylabel("# of words")
plt.title("# of words per line")
plt.show()
        
```

```python
name = ['Mike', 'Elaine', 'Ann', 'John', 'Tommy']
height = [173, 168, 171, 189, 179]
weight = [65.4, 59.2, 63.6, 88.4, 68.7]
gender = ['M', 'F', 'F', 'M','M']

plt.scatter(weight, height, label=name)
plt.legend()

plt.xlabel("height")
plt.ylabel("weight")
```

```python
import pandas as pd

x = pd.DataFrame({"height":height, "weight":weight, "gender":gender}, name)
x
```

```python
x['bmi'] = ((x.weight / x.height) ** 2) * 100
x
```

```python
print(x[x.gender == "M"]['bmi'].mean())
print(x[x.gender == "F"]['weight'].mean())
```

```python
ages = [2, 18, 30, 3, 51, 62, 44, 9, 22, 12, 26, 38, 55, 68, 88, 28, 14]
groupnames = {'Young', 'Adult', 'Old'}
cats = pd.cut(ages, [0, 20, 50, 100], labels=groupnames)
pd.value_counts(cats)
```
