```python
import re

a = "010-1234-5678, 010-123-45678, 031-8041-0010, 0101-123-1231, 0101-1213-11231, 010-123123-1234"

pat = re.compile("(?<!\d)\d{3}-\d{3,4}-\d{4}(?!\d)")
pat.findall(a)
```

```python
data = """010913-3456789 Hong
010913-5456789 Kim  010913-5456789  Park
910913-1456789 Lee"""

pat = re.compile("^(?<!\d)\d{6}-[56]\d{6}(?!\d)\s+\w+", re.M)
pat.findall(data)
```

```python
import time

a = ['s', 'd', 'h', 'c']
b = ['a', '2', '3', '4', '5', '6', '7', '8', '9', '10', 'j', 'q', 'k']
np.random.seed(int(time.time()))
print(np.random.choice([i + j for i in a for j in b], 7, replace=False))
```

```python
import matplotlib.pyplot as plt

x = np.arange(-10, 20)
plt.plot(x, 2*x-5, 'r--')
plt.plot(x, x)
```

```python
x = np.arange(0, 10, 0.1)

plt.plot(x, np.e**(x - 2) + 2) 
plt.plot(x, (x - 2)**2)

plt.xlabel("height")
plt.ylabel("weight")
plt.title("relation")

plt.axis([-2, 11, -20, 20])

plt.show()
```
