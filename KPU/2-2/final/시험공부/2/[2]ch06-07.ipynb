```python
def mulTable(n):
    return [i * n for i in range(1, 10)]

mulTable(2)
```

```python
def getTotalPage(m, n):
    return (m + n - 1) // n

getTotalPage(30, 10)
```

```python
import os

def search(dirname):
    filenames = os.listdir(dirname)
    for filename in filenames:
        full_filename = os.path.join(dirname, filename)
        if os.path.isdir(full_filename):
            search(full_filename)
        else:
            ext = os.path.splitext(full_filename)[-1]
            if ext == '.py':
                print(full_filename)
```

```python
for (path, dir, files) in os.walk("c:/"):
    for filename in files:
        ext = os.path.splitext(filename)[-1]
        if ext == '.py':
            print('%s%s' % (path, filename))
```

```python
import re
```

```python
pat = re.compile("(?P<front>\d{6})-(?P<back>\d{7})")
print(pat.sub("\g<front>-*******", """
park 800905-1049118
kim  700905-1059119
"""))
```

```python
print(re.findall("(?<!\d)\d{3}-\d{4}(?!\d)", "123-1234"))
```

```python
p = re.compile("\.")
p.search("3")
```

```python
p = re.compile(r"(0[1-9]|1[12])-([012][0-9]|3[01])")
for i in p.finditer("11-13/2019 12-14/2000"): print(i.group())

p = re.compile('(life|sdfsd)/is/too/short')
result = p.findall("life/is/too/short")
print(result)
```

```python
p = re.compile(r"(?P<temp>\w+)\s+\d+[-]\d+[-]\d+")
m = p.search("park 010-1234-1234")
print(m.group(1))   
print(m.group("temp"))
```

```python
re.search("\w+(?=:)", "http://google.com")
```

```python
re.match(".*\.(?!bat|exe).*$", "asdf.exe")
```

```python
p = re.compile('(blue|white|red)')
p.sub('colour', 'blue socks and red shoes')
```

```python
p = re.compile("(?P<name>\w+)\s+(\d+-\d+-\d+)")
print(p.sub("\g<name>: \g<2>", "12312312 park 123-1234-1234 sasfd 11-11-11"))
```
