```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
plt.plot(x)
plt.show()
```

```python
x = ['a', 'bsdf', 'esdf', 'd']
plt.plot(x, x, label="dfhs")

plt.xlabel("asdf")
plt.ylabel("asasdf")
plt.title("adsf")
plt.legend(loc="upper right")

plt.show()
```

```python
a = np.arange(10, 20, 3)
b = 2 ** (a - 2)

plt.plot(a, b, marker="o", linestyle="--", markersize=5, linewidth=10, alpha=0.5)
```
