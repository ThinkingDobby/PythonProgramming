## 강의자료

```python
import numpy as np
```

```python
print(np.zeros(5))
print(np.mean([1, 2, 3, 4, 5])) # numpy 배열이 아닌 경우에도 평균 계산 가능

a = np.array([1, 2, 3, 4, 5, 6])
print(a.shape)  # 1차원 배열 - (열,)
print(a.mean())

a = np.array([1, 2, 3, 4]).reshape(1, 4)
print(a.shape)
print(a.mean())

a = np.array([[2], [1], [3]])
print(a.shape)
```

```python
x = np.ones((3, 4), int)
print(x)
print(x.shape)
```

```python
x = np.array(np.arange(1, 21)).reshape(4, 5)
# np.array(1, 10, 0.1) 형태로 arange에는 간격 설정 가능
print(x)
print(x[:][3])  # x[:]의 [3] 원소 -> x 전체에서 4번째 원소
print(x[:, 1])  # x[:, 1] -> 전체 행, 두번째 열에 해당되는 모든 원소
print(x[:, 1:2])
print(x[1:5:2, 2:4])  # 2에서 5행(2스텝), 3에서 4열에 해당되는 모든 원소
print(x[1, 1])
x
```

```python
import math

print(math.sqrt(2))
# math.sqrt([2, 4, 9])

print(np.sqrt(2))
print(np.sqrt([4, 9, 16]))
```

```python
x = np.zeros([4, 3])
print(x)
x[1:3, 1] = 1
x += 3
print(x)
```

```python
a = np.array([[1, 2],[3, 4]])
print(a)
print(a.transpose())

print(np.transpose([[1, 2],[3, 4]])) # 파이썬 리스트에 사용
```

```python
x = range(10)
y = np.array([i for i in range(0, 20, 2)])
print(x)
print(y)
print(x + y)    # 원소 개수 같아야 가능
```

```python
a = np.array([i * 10 for i in range(1, 8)])
print(a)

print(a[[1, 3, 5]])
# print(a[(1, 3, 5)])   # 여러 인덱스 한번에 접근 시 반드시 리스트 사용

i = [1, 2]
print(a[i])
```

```python
a = np.array([10, 20, 30, 40, 50, 60, 70])
b = a[1:6:2]
print(type(b))
print(b)
b[1] = 10
print(b)
print(type(a))
print(type(b))
```

```python
a = np.array([10, 20, 30, 40, 50, 60, 70])
print(a > 50)
print(a[a > 50])
```

```python
import random

print(random.random())
print(np.random.rand())
print(np.random.rand(5))

print(random.randint(1, 5)) # 1..5
print(np.random.randint(1, 5, (2, 3)))  # 1 until 5

# randint, random, standard_normal -> 크기를 하나의 인자로 지정(다차원인 경우 튜플로)
# rand, randn -> 크기를 인자 여러개로 지정
print(np.random.random((2, 3)))
# np.random.random(2, 3)
# np.random.rand((2, 3))
print(np.random.rand(2, 3))
```

```python
a = np.random.random(5)
print(a)
print(np.all(a > 0.3))
print(np.any(a > 0.8))
```

```python
print(np.linspace(0, 12, 2, dtype=int))
print(np.linspace(0, 12, 4, retstep=True))  # 간격 출력
print(np.linspace(0, 12, 4, endpoint=False))
```

```python
np.logspace(np.log10(10), np.log10(100), 5)
```
