```python
p = re.compile("^python\s\w+", re.M)

data = """python one
life is too short
python two
you need python
python three"""

p.findall(data)
```

```python
p = re.compile('(life|sdfsd)/is/too/short')
result = p.findall("life/is/too/short")
print(result)   # findall로 탐색 시 파이프, 그룹핑 잘 적용되지 않는 문제
```

```python
p = re.compile(r'\bclass\b')
print(p.search('no class at all'))

p = re.compile('\sclass\s')
print(p.search('no class at all'))
```

```python
p = re.compile(r'\Bclass\B')
p.search('the declassified algorithm')
```

```python
p = re.compile('(ABC)+')
m = p.search('ABCABCABC OK?')
print(m)
```

```python
p = re.compile(r"(\w+)\s+((\d+)[-]\d+[-]\d+)")
m = p.search("park 010-1234-1234")
print(m)
print(m.group())
print(m.group(1))
print(m.group(2))
print(m.group(3))
```

```python
p = re.compile(r'(\b\w+)\s+\1')
p.search('Paris in the the spring').group() # 첫번째 그룹에 해당하는 문자열이 the인 경우 \1도 the가 됨
```

```python
p = re.compile("(.+):")
m = p.match("http://google.com")
print(m.group(1))

p = re.compile(".+(?=:)")
m = p.match("http://google.com")
print(m.group())
```

```python
p = re.compile('.*[.](?!bat$).*$')
m = p.search("adfasfd.exe")
print(m)
```

```python
p = re.compile('(blue|white|red)')
print(p.sub('colour', 'blue socks and red shoes', count=1))

p = re.compile('(blue|white|red)')
print(p.subn('colour', 'blue socks and red shoes'))
```

```python
p = re.compile('(?P<name>\w+)\s+(?P<phone>\d+-\d+-\d+)')
m = p.sub('\g<phone> \g<name>', "123 park 010-1234-1234")   # <>내에는 따옴표x
print(m)    # 패턴에 매치되는 부분만 변해서 나오고 나머지는 유지

m = p.match("park 010-1234-5678")
print(m.group("name"))
print(m.group(1))   # 이름 지정해도 여전히 인덱스 사용 가능
```

```python
def hexrepl(match):
    value = int(match.group())
    return hex(value)

p = re.compile('\d+')
p.sub(hexrepl, 'Call 65490 for printing, 49152 for user code.')
```

```python
s = '<html><head><title>Title</title>'
print(re.match('<.*>', s).group())
print(re.match('<.*?>', s).group())
```

## 수업내용

```python
import re

data = """
park 800905-1049118
kim  700905-1059119
"""

p = re.compile("\d{6}-\d{7}")
p.findall(data)
```

```python
p = re.compile("(?<!\d)0\d{2}-\d{3,4}-\d{4}(?!\d)")
m = p.findall("010-1234-5678, 010-123-45678, 031-8041-0010, 123-456-7890")
print(m)
```
