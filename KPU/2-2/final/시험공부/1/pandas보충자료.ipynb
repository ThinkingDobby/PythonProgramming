```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
a = pd.Series([4, 7, -5, 3])
print(a.index)
print(a.values) # numpy 배열
type(a.values)

a = pd.Series(a.values, ['d', 'b', 'a', 'c'])
print(a.index)
print(a['b'])

print('a' in a) # in은 인덱스에 대해 조사
```

```python
print(pd.isnull(a))
print(pd.notnull(a))
```

```python
sdata = {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}
obj3 = pd.Series(sdata)
print(obj3)

print(pd.Series(sdata).reindex(['California', 'Ohio', 'Oregon', 'Texas']))

# 딕셔너리로 Series 만들면 key들이 index가 됨, 이 때 index 별도 지정해주면 reindex 처럼 작동
states = ['California', 'Ohio', 'Oregon', 'Texas']
obj4 = pd.Series(sdata, index=states)
print(obj4)
```

```python
data = {'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada', 'Nevada'],
        'year': [2000, 2001, 2002, 2001, 2002, 2003],
        'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}
frame = pd.DataFrame(data)
display(frame)
"state" in frame        # in - 데이터프레임에서는 컬럼에 대해 조사
```

```python
x = pd.DataFrame(data, columns=['year', 'state', 'pop'], index=['one', 'two', 'three', 'four',
                             'five', 'six'])    # 딕셔너리로 만들며 컬럼 설정 -> 컬럼 위치 재설정
display(x)

print(x[['state', 'pop', 'year']])
x = x[['state', 'pop', 'year']] # 컬럼 위치 교환
display(x)
x = x.reindex(columns=['state', 'year', 'pop'])
display(x)

# x.columns = ['year', 'state', 'pop']  # 컬럼 이름만 변경
print(x.columns)
print(x.index)
display(x)
```

```python
print(x.iloc[3])    # 행을 인덱스 숫자로 접근
print(x.loc['four'])    # 행을 인덱스 이름으로 접근
```

```python
a = pd.Series([1.2, 1.5, 1.7], index=['two', 'four', 'five'])
display(x)
x['pop'] = a  # 시리즈를 이용해 특정 컬럼 값 지정도 가능
display(x)
```

```python
x.loc['two'] = ['asdf', 1.1, 2000]
display(x)
```

```python
x['a'] = (x['year'] == 2000)    # 새 컬럼 추가
display(x)
del x['a']
display(x)  # 컬럼 제거
```

```python
obj = pd.Series(np.arange(4.), index=['a', 'b', 'd', 'c'])
print(obj)
print(obj['b':'c'])
```

```python
a = pd.DataFrame(np.arange(16).reshape(4, 4), index=[1, 2, 3, 4], columns=['a', 'b', 'c', 'd'])
display(a)

x = lambda x: x + 3
display(a.apply(x)) # 열 단위 적용 (이 경우는 요소 직접 적용과 같음(열 + 3이 브로드캐스팅됨))

x = lambda x: '%.2f' % x
a.applymap(x)  # 요소별로 함수 직접 적용
```

```python
frame = pd.DataFrame(np.arange(8).reshape((2, 4)),
                     index=['three', 'one'],
                     columns=['d', 'a', 'b', 'c'])
display(frame)
display(frame.sort_index()) # 인덱스로 정렬, 객체 자체에는 적용 x
display(frame.sort_index(axis=1))   # 컬럼으로 정렬
```
