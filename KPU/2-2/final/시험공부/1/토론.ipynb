7장RE (1), (2) 한국전화번호 매칭

```python
import re

a = "010-1234-5678, 010-123-45678, 031-8041-0010, 0101-123-1231, 010-123-1231, 010-123123-1234"

p = re.compile("(?<!\d)0\d{2}-\d{3,4}-\d{4}(?!\d)")
# 모든 경우 다 걸러내기 어려움 - 예제 위주로 걸러낼 것
p.findall(a)
```

7장RE (3) 주민번호 중 외국사람 찾아내기

```python
data = """010913-3456789 Hong
010913-5456789 Kim  010913-5456789  Park
910913-1456789 Lee"""

p = re.compile("^\d{6}-[56]\d{6}", re.M)
p.findall(data)
```

gplotlib (2)

```python
x = np.arange(0, 10, 0.1)

plt.plot(x, np.e**(x - 2) + 2)
plt.plot(x, (x - 2)**2)

plt.xlabel("height")
plt.ylabel("weight")
plt.title("relation of height and weight")

plt.axis([-2, 11, -20, 20])

plt.show()
```
