```python
class FourCal:
    x = 10
    def setdata(self, a, b):
        self.x = a  # 주석 달면 마지막 명령 결과 10으로 나옴
        self.y = b
    def add(self):
        return self.x + self.y
    def sub(self):
        return self.x - self.y
    def mul(self):
        return self.x * self.y
    def div(self):
        return self.x / self.y

a = FourCal();
a.setdata(2, 3)
print(FourCal.mul(a))
print(a.add())
print(a.x)  # 객체명.필드로 접근 시 인스턴스 영역 -> 클래스 영역 순으로 탐색
```

```python
class FourCal:
    def __init__(self, a, b):
        self.x = a
        self.y = b
    def add(self):
        return self.x + self.y
    def sub(self):
        return self.x - self.y
    def mul(self):
        return self.x * self.y
    def div(self):
        return self.x / self.y

a = FourCal(2, 3);
print(FourCal.mul(a))
print(a.add())
print(a.x)
```

```python
class MoreFourCal(FourCal):
    def __init__(self, a, b):
        super().__init__(a, b)
        print(a, b)
    def pow(self):
        return self.x ** self.y
    def div(self):
        if self.y == 0: return 0
        else: return super().div()

b = MoreFourCal(5, 4)
print(b.div())
print(b.pow())

c = MoreFourCal(4, 0)
print(c.div())
```

### 연습문제

```python
class Calculator:
    def __init__(self):
        self.value = 0

    def add(self, val):
        self.value += val

class UpgradeCalculator(Calculator):
    def minus(self, val):
        self.value -= val

cal = UpgradeCalculator()
cal.add(10)
cal.minus(7)

print(cal.value)
```

```python
class MaxLimitCalculator(Calculator):
    def add(self, val):
        super().add(val)
        if self.value > 100: self.value = 100

cal = MaxLimitCalculator()
cal.add(50)
cal.add(60)

print(cal.value)
```
