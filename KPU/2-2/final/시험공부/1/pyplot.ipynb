## 강의자료

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
plt.plot(x)
plt.show()
```

```python
plt.plot([1]) # default: 실선 - 항목 하나일 때는 연결 x - 보이지 x
```

```python
x = [13, 15, 16, 17, 18]
plt.plot(range(len(x)), x)
plt.show()
```

```python
plt.plot(["a", "b", "c", "d"], [10, 2, 3, 4], label="Asdf")
plt.legend()
plt.show()
```

```python
x = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
y = [13, 16, 15, 18, 16, 17, 16]

plt.plot(x, y)
plt.xlabel("Daily")
plt.ylabel("No. Of Books Sold")

plt.show()
```

```python
x = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
y = [13, 16, 15, 18, 16, 17, 16]
z = [17, 14, 17, 16, 15, 17, 16]

plt.plot(x, y, label="Sold")
plt.plot(x, z, label="asdf")
plt.legend(loc="upper left")

plt.xlabel("x")
plt.ylabel("y")

plt.title("Leadsf")

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, 'c--o')   # 색, 점, 선 지정 가능
# 점, 선 한번에 지정 가능

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 30]

plt.axis([0, 6, 0, 30]) # default: 최소 ~ 최대
plt.plot(x, y, 'o')
```

```python
p = np.array([[1, 2], [2, 3], [4, 2], [5, 4]])
plt.plot(p[:, 0], p[:, 1], 'ro')

n = np.array([3, 2.5])
plt.plot(n[0], n[1], 'bo', markersize=10)
```

```python
x = np.linspace(0, 5, 10)
y = x ** 2 

plt.plot(x, y, 'bs-', linewidth=2, markersize=5, color="red")
```

```python
x = np.linspace(0, 10, 20)
y1 = x**2.0
y2 = x**1.5

plt.plot(x, y1, 'bo-', markersize=10, label="First")
plt.plot(x, y2, 'gs-', markersize=10, label="Second")

plt.xlabel('x')
plt.ylabel('y')

plt.axis([-0.5, 10.5, -5, 105])

plt.legend(loc="upper left")

plt.show()
```

```python
x = np.random.standard_normal(1000)
y = np.random.randn(1000)

plt.hist(x)
plt.hist(y)
```

```python
x = np.random.standard_normal(1000)

plt.hist(x, density=True)
```

```python
x = np.random.rand(30)
y = np.random.rand(30)
colors = np.random.rand(30)
shape = np.pi * (np.random.rand(30) * 20)**2 # 0~19 사이 30개 값을 제곱한 것에 파이 곱함
plt.scatter(x, y, s=shape, c=colors, marker="*", alpha=0.7)
# 스캐터 차트 이용하면 각각의 x, y 포인트에 대한 모양과 색 지정 가능
```

## 수업내용

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(6)
y = np.array([1, 12, 5, 19, 22, 7])

plt.plot(x, y, '-->r', label="tri")
plt.legend(loc="upper left")

plt.show()
```

```python
x = np.arange(-10, 21)
y = x * 2 - 5
plt.plot(x, y, "r--")
plt.plot(x, x, "b")

plt.show()
```

```python
x = np.arange(0, 11, 0.1)

plt.plot(x, np.e ** (x - 2) + 2)
plt.plot(x, (x - 2)**2)

plt.scatter(x, np.e ** (x - 2) + 2, c = np.random.rand(len(x)))
plt.scatter(x, (x - 2)**2, c = np.random.rand(len(x)))

plt.axis([-2, 11, -20, 20])
plt.title("relation")

plt.xlabel("height")
plt.ylabel("weight") 
```

```python
plt.plot([[1, 2], [6, -1], [3, 4]])
# x가 0일 때 1, 1일 때 6, 2일 때 3 짜리 하나
# x가 0일 때 2, 1일 때 -1, 2일 때 4 짜리 하나
```

```python
x = np.linspace(0, 5, 10)
print(x)
y = x ** 2
plt.plot(x, y)

x = np.arange(0, 5, 0.1)
print(x)
y = x**2
plt.plot(x, y)
```
