## 수업내용

```python
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
x = pd.Series(np.arange(10, 20, 2), index = ['서울', '부산', '광주', '대전', '대구'])
print(x)

y = x.reindex(['부산', '광주', '서울', '용인', '대전'])
print(y)  # NaN이 추가되면서 타입이 실수형으로 바뀜
# print(y.astype(int))    # NaN은 정수형으로 나타낼 수 없음
```

```python
x = pd.Series([1, 2, 3, 4])
print(x.values)
print(type(x.values))   # numpy 배열로 반환됨, 인덱스는 x

print(x.index > 1)
print(x.values > 2)
print(x[x.values > 2])
```

```python
x = pd.Series([7, 3, 6, 8], ['서울', '대구', '부산', '광주'])
print(x.index + "a") # Index 객체에도 연산 브로드캐스트 가능
```

```python
# 요약: 인덱스에 대한 bfill, ffill은 사전순으로 적용되는 것

print("----1----")
x = pd.Series([1, 2, 3, 4], ['a', 'b', 'c', 'd'])   # 인덱스는 사전순이어야(사전순이 아니면 method 적용 불가)
print(x.reindex(['e', 'b', 'a', 'e'], method='bfill')) # NaN 그대로
# 문자 인덱스 'e'의 경우 뒷 인덱스와 값이 존재하지 않기 때문에 NaN 유지
print(x.reindex(['e', 'b', 'a', 'e', 'f'], method='ffill')) # ffill, pad: 앞 인덱스에 해당하는 value 가져옴
# 문자 인덱스 'e'인 경우 앞 인덱스 'd'의 값을 가져오므로 4가 들어가게됨

print("----2----") 
print(x.reindex(['g', 'a', 'b', 'c', 'd', 'e', 'f'], method='ffill'))  # 이전 인덱스의 값 가져옴
```

```python
print("----3----")
x = pd.Series([8, 2, 3, 4], ['aa', 'bb', 'dd', 'ee'])   # 인덱스가 사전 순이 아닌 경우 오류 발생
print(x.reindex(['sdf', 'aa', 'bb', 'cc', 'dd', 'ee', 'sd']))
print(x.reindex(['sdf', 'aa', 'bb', 'cc', 'dd', 'ee', 'sd'], method="ffill"))   # 사전 순으로 "sdf" 앞의 가장 가까운 인덱스는 "ee"이므로 ee의 값 불러옴

print("----4----")
x = pd.Series(['a', 'b', 'c', 'd'], [2, 6, 7, 8])
print(x)
x.reindex(np.arange(0, 9), method="ffill")

# 인덱스 0일 때의 ffill -> NaN
# 값이 정해진 마지막 인덱스 이후에 대한 ffill -> NaN
```

```python
x = pd.DataFrame([[1, 2, 3], [2, 3, 4], [4, 5, 6]], [1, 2, 3], columns=['a', 'b', 'c'])
# 2차원 배열 -> 각 행이 행이됨
display(x)

display(x.iloc[0:2, 0:2])
```

```python
data = {'age' : [23, 43, 12, 45], 'name' : ['민준', '현우', '서연', '동현']
# 딕셔너리 -> 각 항의 키가 컬럼명, 값이 컬럼이 됨
, 'height' : [175.3, 180.3, 165.8, 172.7]}
x = pd.DataFrame(data, index=[1, 2, 3, 4])
x
```

```python
x[['age', 'name']]
```

```python
print(x[x.age > 30])   # x.age의 결과가 numpy배열이기 때문에 가능
print(x.age > 30)
print(x[[True, True, False, True]]) # boolean 배열 - 인덱스에 대해 적용
```

```python
print(x.mean())
print(x.mean(1))
```

```python
x = pd.Series(['php', 'python', 'java', 'c#'], [1, 2, 3, 4])
x.apply(lambda a: a.upper())    # 제작한 함수 apply 이용해 값 대상 브로드캐스트 가능
```
