#### 1. 10,000 이하의 정수 중에서 다음 조건을 만족하는 가장 큰 수를 구하시오. (3점)
o 솟수(Prime Numer)이다. 

o 피보나치 넘버이다.

```python
def get_primes(peek):
    check = [True for _ in range(0, peek + 1)]
    primes = []

    check[0] = check[1] = False
    for i in range(2, peek + 1):
        if check[i]:
            primes.append(i)
            for j in range(2 * i, peek + 1, i):
                check[j] = False
    return primes

def get_fibos(peek):
    fibos = []
    a = 0
    b = 1
    while True:
        if a > peek: break
        fibos.append(a)
        a, b = b, a + b
    return fibos

primes = set(get_primes(10000))
fibos = set(get_fibos(10000))
print(sorted(list(primes & fibos))[-1])
```

#### 2. 주어진 영문 문자열을 이용하여 약자를 만드는 프로그램을 작성하시오. (2점)
o 입력: input()함수를 이용하여, 사용자의 영문 문자열을 입력을 받는다.  
o 약자만들기 조건
 - 약자를 추출대상 단어는 길이가 큰 우선으로 최대 4개 추출(동일길이 경우 앞단어 선택)
 - 각단어의 첫 글자를 대문자로 변환하여 구성한다.  
o 기타 다른 조건은 위 조건을 위반하지 않는 범위에서 가정를 함께 코멘트로 작성하고 답안을 작성할 수 있음.

예1) >> Enter FullName: korea polytechnic university 
  
  출력결과:  KPU
  
예2) >> Enter FullName: as soon as possible 
  
  출력결과:  ASAP
  
예4) >> Enter FullName: as far as i know

  출력결과: AFAK  
  
예5) >> Enter FullName: How dare you do that?

  출력결과: HDYT

```python
iv = input("Enter FullName: ").split()
memo = [(i, len(i)) for i in iv]
check = sorted([i[1] for i in memo], reverse=True)[:4]  # 길이가 4보다 길 때만 4개까지 자름, 짧을 때는 상관 x
# : 을 이용해 인덱싱 시 오류 발생 x, 범위 초과 시 -> 해당되는 요소 있으면 포함, 나머지는 무시("")

print("출력결과: ", end='')
for i in iv:
    if len(i) in check:
        print(i[0].upper(), end='')
        check.remove(len(i))
```
