#### 1. 10,000 이하의 정수 중에서 다음 조건을 만족하는 가장 큰 수를 구하시오. (3점)
o 솟수(Prime Numer)이다. 

o 피보나치 넘버이다.

```python
def fib(n):
    a,b = 1,1
    if n==1 or n==2:
        return 1    
    for i in range(1,n):
        a,b = b, a+b
    return a

def isprime(n):
    if n <= 1:
        return False
    for a in range(2,n):
        if n%a == 0:
            return False
    return True

# print(fib(5))

n = 3
while True:
    curfibo=fib(n)
    if curfibo <= 10000:
        if isprime(curfibo): 
            max_fibo = curfibo
        n+=1
    else:
        break
    

print(max_fibo)    
```

#### 2. 주어진 영문 문자열을 이용하여 약자를 만드는 프로그램을 작성하시오. (2점)
o 입력: input()함수를 이용하여, 사용자의 영문 문자열을 입력을 받는다.  
o 약자만들기 조건
 - 약자를 추출대상 단어는 길이가 큰 우선으로 최대 4개 추출(동일길이 경우 앞단어 선택)
 - 각단어의 첫 글자를 대문자로 변환하여 구성한다.  
o 기타 다른 조건은 위 조건을 위반하지 않는 범위에서 가정를 함께 코멘트로 작성하고 답안을 작성할 수 있음.

예1) >> Enter FullName: korea polytechnic university 
  
  출력결과:  KPU
  
예2) >> Enter FullName: as soon as possible 
  
  출력결과:  ASAP
  
예4) >> Enter FullName: as far as i know

  출력결과: AFAK  
  
예5) >> Enter FullName: How dare you do that?

  출력결과: HDYT

```python
name = input(">> Enter FullName: ").split()

word = sorted(name, key=len, reverse=True)  # key=함수 - len, sum 등 사용 가능
if len(word) >= 4: word = word[:4]

result = ''
for n in name:
    if word and n in word:
        result += n[0].upper()
print(result)

```

#### 4. 회문을 찾는 함수를 만들고, 이를 이용하여 input.txt 파일에서 회문을 찾아 출력하는 프로그램을 작성하시오. (단, 대소문자는 구분하지 않는다) (3점)
o 시험중 공지: 한글자 짜리는 포함하지 않음.

(참고) 회문이란, 역순으로 해도 동일한 결과를 만들어 내는 단어 

    - 예) Mom Madam a2332a Dad 123421      

```python
# 추가채점반영사항: 구분자 명시하지 않아, white space만 사용하는 경우 가점 부여
#                   실행시, radar 만 나온 경우도 50% 점수를 반영

def ispalindrome(s):
    return s == s[::-1]

f1 = open("input.txt", 'r')

c = f1.read()
words = c.lower().split()

for w in words:
    w2 = ''.join([i for i in w if i.isalpha()])
    if ispalindrome(w2) and len(w2) != 1: 
        print(w)        
f1.close()


""" Sol 2: readline으로 처리하는 경우
#  화이트스페이스, 알파벳이외 문자를 구분자로 활용하는 경우: 모두 4개 출력
while True:
    line = f1.readline()
    if not line: break
        
    for s in line.lower().split():
        s = ''.join([i for i in s if i.isalpha()]) # 특수문자 제외, 알파벳만
        if len(s) != 1 and ispalindrome(s):
            print(s)
"""            
```
