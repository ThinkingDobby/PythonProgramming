```python
class FourCal:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    # def setdata(self, a, b):
    #     self.a = a
    #     self.b = b

    def add(self):
        return self.a + self.b
    
    def mul(self):
        return self.a * self.b
    
    def sub(self):
        return self.a - self.b

    def div(self):
            return self.a / self.b

a = FourCal(4, 2)
# a.setdata(4, 2) # a.a에 4, a.b에 2 저장
print(a.a, a.b)
print(a.add())
print(a.mul())
print(a.sub())
print(a.div())

# b = FourCal()
# FourCal.setdata(b, 3, 5)    # 클래스를 통해 메서드 호출도 가능
# print(FourCal.add(b))   # self 명시 필요

class MoreFourCal(FourCal):
    def pow(self):
        return self.a ** self.b

    def div(self):
        if self.b != 0:
            return self.a / self.b  
        else:
            return 0

c = MoreFourCal(2, 0)
print(c.add())
print(c.pow())
print(c.div())
```
