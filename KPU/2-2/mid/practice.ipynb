```python
def f(n):
    if n < 2:
        return n
    else:
        return f(n - 2) + f(n - 1)

def s(n):
    i = 1
    a, b = 0, 1
    while i < n:
        a, b = b, a + b
        i += 1
    return b

n = int(input())
print(f(n))
print(s(n))
```
