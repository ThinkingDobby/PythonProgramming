```python
def get_primes(l):
    pl = []
    cl = [True for i in range(0, l + 1)]
    cl[0] = cl[1] = False
    for i in range(2, l + 1):
        if cl[i]:
            pl.append(i)
            for j in range(i * 2, l + 1, i):
                cl[j] = False
    return pl

def get_fibos(l):
    fl = []
    a = 0
    b = 1
    while True:
        if a > l: break
        fl.append(a)
        a, b = b, a + b
    return fl
    
print(sorted(list(set(get_fibos(10000)) & set(get_primes(10000))))[-1])
```

```python
il = input().split()
memo = sorted([len(i) for i in il], reverse=True)[:4]

ans = ""
for i in il:
    if len(i) in memo:
        memo.remove(len(i))
        ans += i[0].upper()
        if not memo: break

print(ans)
```

### 과제

```python
## 5.1 클래스
# 다음 Airforce 클래스를 이용하여 질문에 답하라. 
class Airforce:
    def __init__(self, name, p):
        self.name = name
        self.power = p
```

```python
a1 = Airforce("F14", 50)
print(a1.name + ", " + str(a1.power))
```

```python
class Jet(Airforce):
    cnt = 0
    def __init__(self, name, p):
        super().__init__(name, p)   # self의 cnt가 존재 x - 클래스 변수의 값이 변할 것 같지만,
        self.cnt += 1   # self.cnt(인스턴스 변수에 저장됨) = self.cnt(클래스 변수 들어옴) + 1
    def __del__(self):  # 따라서 결론적으로 클래스 변수의 값은 변하지 x
        self.cnt -= 1

j1 = Jet("F14",50)
print(Jet.cnt)
j2 = Jet("SU33", 60)
j3 = Jet("Mirage2000", 45)
del j2
print("# of Jets=", Jet.cnt)
```

```python
class Test:
    t = 5
    def a(self):
        return self.t
    def b(self, n):
        self.t += n

v = Test()
print(v.a())

v.b(3)
print(v.a())

print(Test.t)
```

```python
class Helicopter(Airforce):
    def __str__(self):
        return "(%s, %d)" % (self.name, self.power)

h =  Helicopter("F16",150)   
print(h)
```
