```python
def get_divisors(n):
    a = []
    i = 1
    while i * i <= n:
        if n % i == 0:
            if n // i == i: a.append(i)
            else:
                a.append(i)
                a.append(n // i)
        i += 1
    return sorted(a)

print(get_divisors(100))
```

```python
name = ['Mike', 'Elaine', 'Ann', 'John', 'Tommy']
height = [173, 168, 171, 189, 179]
weight = [65.4, 59.2, 63.6, 88.4, 68.7]
gender = ['M', 'F', 'F', 'M','M']

tmp = list(map(lambda x: list(x), zip(height, weight, gender)))
ans = {name[i]:[height[i], weight[i], gender[i]] for i in range(len(name))}
ans2 = {name[i]:tmp[i] for i in range(len(name))}
print(ans)
print(ans2)
```

```python
data_a = "Let me find Palindrome in the following words. Father is dad and mother is mom. Brother is male and sister is female. All of us like kayak and pop song."
d= dict()
def words_sorted(data):
    
    words = data.split()

    for word in words:
        if word in d:
            d[word] = d[word] +1
        else:
            d[word] = 1


    result = sorted(d.items())
    return result

print(sorted(words_sorted(data_a), key=lambda a: a[1]))
```
