```python
while True:
    try:
        f_name = input("파일 패스 입력: ")
        f = open(f_name, 'r', encoding = "UTF-8")
    except FileNotFoundError:
        print(f_name,"파일 오픈 실패")
        continue
    else:
        print(f_name, "파일 오픈 성공")
        break

f_words = f.read().split()
f_dic = {n:0 for n in f_words}  # 딕셔너리 키는 중복 값 있을 시 덮어 쓰여짐
for i in f_words:
    f_dic[i] += 1

s_f_dic = sorted(f_dic.items(), key = lambda a: a[1], reverse = True)

print(f_name, "파일의 단어 수:", len(f_words))
print("최빈 단어와 출현회수:", s_f_dic[0][0], s_f_dic[0][1])
print("차빈 단어와 출현회수:", s_f_dic[1][0], s_f_dic[1][1])
print("회문: ", end = '')
print(*[i for i in f_words if i == i[::-1]])
```
