```python
def get_primes(n):
    check = [True for _ in range(0, n + 1)]
    primes = []

    check[0] = check[1] = False
    for i in range(2, n + 1):
        if check[i]:
            primes.append(i)
            for j in range(2 * i, n + 1, i):
                check[j] = False
    return primes

def get_fibo(cnt, a, b):
    if cnt == 0:
        return a
    else:
        return get_fibo(cnt - 1, b, a + b)

MAX = 100000000

fibos = []
for i in range(0, MAX):
    tmp = get_fibo(i, 0, 1)
    if tmp > MAX: break
    fibos.append(tmp)

primes = get_primes(MAX)
ans_list = list(set(fibos) & set(primes))
ans_list.sort(reverse=True)

while True:
    n = int(input("정수입력(1억이하): "))
    if n < 0:
        print("<종료>")
        break
    print("입력받은 정수: %d" % n)

    tmp = n
    for i in ans_list:
        if i < n:
            tmp = i
            break
    
    print("%d이하 솟수&피보나치 최대수: %d" % (n, tmp))


```
